```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

import matplotlib.pyplot as plt
```

```python
x = range(0,9)
print(x)

y = [1,2,3,4,3,5,4,3,2,]
plt.bar(x,y,color=["purple","red","green"])
plt.xlabel("X LABEL")
plt.ylabel("Y LABEL")
plt.show()
```

```python
z = [1,2,3,4,5]
type = ["Cycle","Bike","Scooter","Van","Car"]
plt.pie(z,labels=type)

plt.show()
```

```python
cars = pd.read_csv("mtcars.csv")
columns = cars.columns
cols = list(columns)
cols
cols[0] = "car_names"
print(cols)

```

```python
cars.columns = cols
cars.head()
```

```python
mpg = cars["mpg"]
mpg
```

```python

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names,rotation=60,fontsize="medium")
ax.set_title( "Miles per Gallon of Cars in mtcars")
ax.set_xlabel("Car Names")
ax.set_ylabel("Number of Gallons")
ax.plot(mpg)
plt.show()
```
