```python
import pandas as pd
import numpy as np
```

```python
cars = pd.read_csv("mtcars.csv")
```

```python
cars.head()
```

```python
carb = cars["carb"]
carb.value_counts()
```

```python
cars_subset = cars[["cyl","vs","am","gear","carb"]]
cars_subset.head()
```

```python
carb_group = cars.groupby(cars["carb"])
carb_group.mean()
```

```python
gears_group = cars_subset.groupby(cars_subset["gear"])
gears_group.describe()
```

```python
#creating a categorical variable - dtype = "category"
cars["group"] = pd.Series(cars["gear"],dtype="category") #creates a new column
cars.head()
```
