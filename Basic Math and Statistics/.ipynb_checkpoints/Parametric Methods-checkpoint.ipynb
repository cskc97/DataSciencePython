Correlation does not imply causation

Pearson Correlation Coefficient for two variables
R = 1 -> Strong positive relationship 
R = 0 -> Not linearly correlated
R = -1 -> Strong negative relationship

Assumptions for the pearson correlation - 
Data is normally distributed, variables are continuous and are linearly related. 

So, if R value is 0 and you haven't tested for the assumptions, the variables "may" be non-linearly correlated also. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from pylab import rcParams
import scipy
from scipy.stats.stats import pearsonr
```

```python
cars = pd.read_csv("mtcars.csv")
cars.head()
```

```python
sb.pairplot(cars)
plt.show()
```

```python
X = cars[["mpg","hp","qsec","wt"]]
sb.pairplot(X)
plt.show()
```

Normally distributed variables are shaped like a bell curve. 
Continuous variables are not discrete. So, the variables have values only at certain discrete points. 


```python
#Using scipy to calculate the Pearson correlation coefficient. 

mpg = cars["mpg"]
hp = cars["hp"]
qsec = cars["qsec"]
wt = cars["wt"]

pearsonr_coefficient, p_value = pearsonr(mpg,hp) #first value is coefficient, and the second is the p_value
pearsonr_coefficient
```

```python
pearsonr_coefficient, p_value = pearsonr(mpg,qsec) #first value is coefficient, and the second is the p_value
pearsonr_coefficient
```

```python
pearsonr_coefficient, p_value = pearsonr(mpg,wt) #first value is coefficient, and the second is the p_value
pearsonr_coefficient
```

mpg,wt variable pair has the strongest correlation when compared to the other correlations. Some machine learning models will require variable pairs that have strong correlations. 

```python
corr = X.corr()
corr #produces pearson coefficient values. 
```

```python
sb.heatmap(corr) 
plt.show() #this can be used to analyze correlations graphically.
```
