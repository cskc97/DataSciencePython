These kind of methods are for non-linear, and even discrete pair variables. 

Spearman's Rank Correlation:
Takes a variable-pair of ordinal data types, finds the R correlation and ranks according to the strength of the correlation. 
Ordinal variable is a numeric variable that can be categorized. 

R = 1 -> Strong positive relationship
R = 0 -> Not correlated
R = -1 -> Strong negative relationship

Assumptions - 
1) Variables are ordinal (so they can be ranked)
2) Variables are related nonlinearly
3) Data is non-normally distributed. 

Chi-Square Tables Test For Independence:

p < 0.05 -> variables are correlated
p > 0.05 -> variables are independent

Assumptions - 
1) variables are categorical or numeric
2) numeric variables have to be binned (breaking up numeric variables in different groups).



```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sb



import scipy
from scipy.stats import spearmanr
from scipy.stats import chi2_contingency
```

```python
cars = pd.read_csv("mtcars.csv")
cars.head()
```

```python
sb.pairplot(cars)
plt.show()
```

```python
X = cars[["cyl","vs","am","gear"]]
sb.pairplot(X)
```

```python
plt.show()
```

```python
''' The data above are all numeric and are discrete. All these values can be ranked into categories, so they are ordinal. 
And a quick glimpse shows they may be related non-linearly. '''
```

```python
cyl = cars["cyl"]
vs = cars["vs"]
am = cars["am"]
gear = cars["gear"]
```

```python
spearmanr(cyl,vs)[0]
```

```python
spearmanr(cyl,am)[0]
```

```python
spearmanr(cyl,gear)[0]
```

```python
#cyl and vs pair show a stronger negative correlation.
```

```python
table = pd.crosstab(cyl,am)
print(table)
print(table.values)
print(type(table.values))
```

```python
chi2, p_value, dof, expected = chi2_contingency(table.values)
p_value
```

```python
# because p_value is <0.05, the variables are correlated. 
```
