```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
DF_obj = DataFrame(np.arange(36).reshape(6,6))
DF_obj
```

```python
DF_obj2 = DataFrame(np.arange(15).reshape(5,3))
DF_obj2
```

```python
# no axis is default and will add more rows. axis = 1 will add more columns.
pd.concat([DF_obj,DF_obj2])
```

```python
pd.concat([DF_obj,DF_obj2],axis=1)
```

```python
DF_obj.drop([0,2]) #will drop the rows 
```

```python
DF_obj
```

```python
DF_obj.drop([0,2],axis=1) #will drop the columns.
```

```python
series_obj = Series(np.arange(6))
series_obj.name="added_variable"
series_obj
```

```python
variable_added = DataFrame.join(DF_obj,series_obj)
variable_added
```

```python
#ignoring index will append without reindexing the row and column series
added_datatable = variable_added.append(variable_added,ignore_index=False)
added_datatable
```

```python
#basically the same thing as concat
DF_obj.append(DF_obj2)
```

```python
#reindexed.
DF_obj.append(DF_obj2,ignore_index=True)
```

```python
# .sort_values(by=[index to sort it by],ascending=[True/False])
DF_obj.sort_values(by=3,ascending=True)
```

```python
DF_obj.sort_values(by=3,ascending=False)
```

```python
DF_obj.sort_values(by=3,ascending=False,axis=1) #sorts by rows. 
```
